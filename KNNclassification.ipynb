```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('heart_failure_clinical_records_dataset.csv')

data.head(5)
```

```python
data.isna().sum()
```

```python
data1= data.drop(['age'],axis=1)
```

```python
deathdiabetes = pd.DataFrame(data.groupby('DEATH_EVENT')['diabetes'].count())
deathdiabetes
```

```python
data1.corr().style.background_gradient()
```

```python
X = data1.iloc[:, :11]
y = data1.iloc[:, 11]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
pd.DataFrame(y_train)
```

```python
pd.DataFrame(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
list_k = []
list_acc_score = []
for k in range (2,18):
    list_k.append(k)
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train,y_train)
    y_pred =  model.predict(X_test)
    score = accuracy_score(y_test,y_pred)
    list_acc_score.append(score)
    print ("With k = %d --> the accuracy is %f" %(k, score))
    print(accuracy_score(y_train, model.predict(X_train)))
datascore = pd.DataFrame(list_k,list_acc_score)
datascore.head()
```
